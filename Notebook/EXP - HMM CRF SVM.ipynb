```python
# Install required libraries
!pip install datasets sklearn_crfsuite seqeval scikit-learn hmmlearn pandas matplotlib seaborn joblib

# Importing necessary Libraries and Dependencies
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn_crfsuite import CRF, metrics
from hmmlearn import hmm
from seqeval.metrics import classification_report, f1_score, precision_score, recall_score
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import itertools
from datasets import load_dataset
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import make_scorer
from tqdm import tqdm

#  Define global configuration parameters for consistency and easy adjustments
class Config:
    RANDOM_STATE = 42
    ALL_TAGS = ['O', 'B-AC', 'I-LF', 'B-LF']
    TAG_MAP = {'O': 0, 'B-AC': 1, 'I-LF': 2, 'B-LF': 3}
    HMM_ITERATIONS = 100
    CRF_ITERATIONS = 100
    DEBUG_MODE = False
    PLOT_RESULTS = True

#  Dataset Analysis and visualize training data: tag distribution, POS tags, and common abbreviations
def analyze_dataset(X_train, y_train, pos_train):
    """Comprehensive dataset visualization and statistics"""
    print("\n Dataset Analysis")

# Flatten and count tag occurrences
    # Tag distribution
    # Plot tag distribution as a bar chart
    tag_counts = pd.Series([tag for seq in y_train for tag in seq]).value_counts()
    tag_names = [Config.ALL_TAGS[i] for i in tag_counts.index]

    plt.figure(figsize=(8,5))
    plt.subplot(1, 2, 1)
    sns.barplot(x=tag_names, y=tag_counts.values)
    plt.title("Tag Distribution")
    plt.xticks(rotation=45)

    # POS distribution
    # Plot top 15 most common POS tags
    pos_counts = pd.Series([pos for seq in pos_train for pos in seq]).value_counts().head(15)
    plt.subplot(1, 2, 2)
    sns.barplot(x=pos_counts.index, y=pos_counts.values)
    plt.title("Top 15 POS Tags")
    plt.xticks(rotation=45)
    plt.tight_layout()
    if Config.PLOT_RESULTS:
        plt.show()

    # Abbreviation analysis
    # Extract and print most frequent abbreviations from B-AC tags
    abbrevs = [X_train[i][j] for i, seq in enumerate(y_train)
               for j, tag in enumerate(seq) if tag == Config.TAG_MAP['B-AC']]
    common_abbrevs = pd.Series(abbrevs).value_counts().head(10)

    print("\nTop 10 Abbreviations:")
    print(common_abbrevs)

    return {
        'tag_distribution': tag_counts,
        'pos_distribution': pos_counts,
        'common_abbreviations': common_abbrevs
    }

# Load and preprocess the PLOD-CW-25 dataset into separate token, POS, and tag sequences

def load_data():
    """Load and preprocess dataset"""
    print("Loading dataset...")
    dataset = load_dataset("surrey-nlp/PLOD-CW-25")
    train_data = dataset["train"]
    test_data = dataset["test"]

    if Config.DEBUG_MODE:
        train_data = train_data.select(range(500))

# Convert string tags into integer indices using TAG_MAP

    X_train = [seq["tokens"] for seq in train_data]
    pos_train = [seq["pos_tags"] for seq in train_data]
    y_train = [[Config.TAG_MAP[tag] for tag in seq["ner_tags"]] for seq in train_data]

    X_test = [seq["tokens"] for seq in test_data]
    pos_test = [seq["pos_tags"] for seq in test_data]
    y_test = [[Config.TAG_MAP[tag] for tag in seq["ner_tags"]] for seq in test_data]

    return X_train, pos_train, y_train, X_test, pos_test, y_test

# Feature Engineering

def word2features(sent, pos, i):
    """Enhanced feature extraction for CRF"""
    word = sent[i]
    # Core features that consistently showed importance:
    features = {
        'bias': 1.0,                        # Helps with class imbalance
        'word.lower()': word.lower(),       # Normalization
        'word[-3:]': word[-3:],             # Capture suffixes
        'word.isupper()': word.isupper(),   # for abbreviations
        'word.istitle()': word.istitle(),
        'word.isdigit()': word.isdigit(),
        'postag': pos[i],                   # POS tags from dataset
        'word.length': len(word),
    }

    # Added context features after initial poor boundary detection
    for j in range(1, 3):                   # Optimal window size from ablation study
        if i > j:
            features.update({
                f'-{j}:word.lower()': sent[i-j].lower(),
                f'-{j}:postag': pos[i-j],
            })
        if i < len(sent)-j:
            features.update({
                f'+{j}:word.lower()': sent[i+j].lower(),
                f'+{j}:postag': pos[i+j],
            })
    return features

def sent2features(sent, pos):
    return [word2features(sent, pos, i) for i in range(len(sent))]

# Model Implementations
# Implementation of Hidden Markov Model using hashed token values as observations

class HMMModel:
    """HMM Implementation - Tradeoff between simplicity and performance"""
    # Used hashing trick to handle vocabulary size
    def __init__(self, n_components, n_iter=50):
        self.model = hmm.CategoricalHMM(
            n_components=n_components,
            n_iter=n_iter,
            random_state=Config.RANDOM_STATE,
            verbose=True
        )

# Convert tokens to hash values for compact representation and prepare data for HMM fitting
    def train(self, X_train, y_train):
        print("Training HMM...")
        X, lengths, y_flat = [], [], []
        for sent, tags in tqdm(zip(X_train, y_train), total=len(X_train)):
            seq = [hash(word) % 1000 for word in sent]
            X.extend(seq)
            lengths.append(len(seq))
            y_flat.extend(tags)

        self.model.fit(np.array(X).reshape(-1, 1), lengths=lengths)

# Predict tags using the trained HMM model and normalize output tag indices
    def predict(self, X_test):
        return [
            [p % len(Config.TAG_MAP) for p in
             self.model.predict(np.array([hash(w) % 1000 for w in sent]).reshape(-1, 1))]
            for sent in X_test
        ]

# Conditional Random Field (CRF) model with hyperparameter tuning using RandomizedSearchCV
class CRFModel:
    """CRF with Hyperparameter Tuning"""
    def __init__(self):
        self.model = None
        self.param_dist = {
            'c1': np.logspace(-3, 2, 10),
            'c2': np.logspace(-3, 2, 10),
            'max_iterations': [50, 100, 200],
        }

# Convert tokens and POS to feature sets and perform randomized hyperparameter tuning
    def train(self, X_train, pos_train, y_train):
        print("Training CRF with hyperparameter tuning...")
        X = [sent2features(s, p) for s, p in zip(X_train, pos_train)]
        y = [[Config.ALL_TAGS[t] for t in seq] for seq in y_train]

        search = RandomizedSearchCV(
            CRF(algorithm='lbfgs', all_possible_transitions=True),
            self.param_dist,
            scoring=make_scorer(metrics.flat_f1_score),
            n_iter=5,
            cv=2,
            verbose=1,
            n_jobs=-1
        )
        search.fit(X, y)
        self.model = search.best_estimator_
        print(f"Best CRF params: {search.best_params_}")

# Predict using trained CRF model and convert labels to integer tags
    def predict(self, X_test, pos_test):
        return [
            [Config.TAG_MAP[p] for p in seq]
            for seq in self.model.predict([sent2features(s, p) for s, p in zip(X_test, pos_test)])
        ]

# Support Vector Machine (SVM) model using contextual TF-IDF features with hyperparameter tuning
class SVMModel:
    """Optimized SVM with Hyperparameter Tuning"""
    def __init__(self):
        self.model = None
        self.vectorizer = TfidfVectorizer(
            ngram_range=(1, 1),
            max_features=2000,
            lowercase=False,
            sublinear_tf=True
        )
        self.param_dist = {
            'C': [0.1, 1.0, 10.0],
            'kernel': ['linear', 'rbf'],
            'gamma': ['scale', 'auto'],
            'class_weight': ['balanced', None]
        }
        self.window_size = 1

# Transform input features using TF-IDF and tune SVM with different kernels and C values
    def train(self, X_train, y_train):
        print("Training SVM with hyperparameter tuning")
        flat_X, flat_y = self._create_window_features(X_train, y_train)
        X = self.vectorizer.fit_transform(flat_X)

        search = RandomizedSearchCV(
            SVC(random_state=Config.RANDOM_STATE),
            self.param_dist,
            cv=2,
            n_iter=5,
            verbose=1,
            n_jobs=-1
        )
        search.fit(X, flat_y)
        self.model = search.best_estimator_
        print(f"Best SVM params: {search.best_params_}")

# Converted surrounding word tokens into a window-based feature representation for each token
    def _create_window_features(self, X_data, y_data):
        flat_X, flat_y = [], []
        for sent, tags in zip(X_data, y_data):
            for i in range(len(sent)):
                context = []
                for j in range(i-self.window_size, i+self.window_size+1):
                    if 0 <= j < len(sent):
                        context.append(sent[j])
                    else:
                        context.append("PADDING")

                features = " ".join(
                    f"{w}_{j-(i-self.window_size)}"
                    for j, w in enumerate(context)
                )
                flat_X.append(features)
                flat_y.append(tags[i])
        return flat_X, flat_y

# Predict token labels by generating context-based features for each token in test data
    def predict(self, X_test):
        predictions = []
        for sent in X_test:
            sent_pred = []
            for i in range(len(sent)):
                context = []
                for j in range(i-self.window_size, i+self.window_size+1):
                    if 0 <= j < len(sent):
                        context.append(sent[j])
                    else:
                        context.append("PADDING")

                features = " ".join(
                    f"{w}_{j-(i-self.window_size)}"
                    for j, w in enumerate(context)
                )
                X = self.vectorizer.transform([features])
                pred = self.model.predict(X)
                sent_pred.append(int(pred[0]))
            predictions.append(sent_pred)
        return predictions

#  Enhanced Error Analysis
# Analyze model performance errors focusing on common mistakes like B-LF/I-LF mislabeling

def error_analysis(X_test, y_true, y_pred, model_name, n_examples=3):
    """ error analysis with examples of  failure patterns observed"""
    print(f"\n {model_name} Error Analysis ")

    # Boundary error detection between B-LF and I-LF
    boundary_errors = []
    for i in range(len(y_true)):
        if len(y_true[i]) != len(y_pred[i]):
            continue

        for j in range(1, len(y_true[i])-1):
            true = y_true[i][j]
            pred = y_pred[i][j]

            # Focus on B-LF/I-LF boundary errors
            if (true == Config.TAG_MAP['B-LF'] and pred == Config.TAG_MAP['I-LF']) or \
               (true == Config.TAG_MAP['I-LF'] and pred == Config.TAG_MAP['B-LF']):
                boundary_errors.append((
                    X_test[i][j-1:j+2],  # Context window
                    y_true[i][j-1:j+2],
                    y_pred[i][j-1:j+2],
                    i, j
                ))

    # Print examples
    print(f"\nFound {len(boundary_errors)} boundary errors")
    if boundary_errors:
        print("\nSample Boundary Errors:")
        for error in boundary_errors[:n_examples]:
            context, true, pred, seq_id, pos = error
            print(f"\nSequence {seq_id}, Position {pos}:")
            print("Context:", " ".join(context))
            print("True:", [Config.ALL_TAGS[t] for t in true])
            print("Pred:", [Config.ALL_TAGS[p] for p in pred])

    # General error examples
    general_errors = []
    for i in range(len(y_true)):
        if len(y_true[i]) != len(y_pred[i]):
            continue

        errors = [(j, y_true[i][j], y_pred[i][j])
                 for j in range(len(y_true[i]))
                 if y_true[i][j] != y_pred[i][j]]

        if errors:
            general_errors.append((i, errors))

    print("\nGeneral Error Examples:")
    for seq_id, errors in general_errors[:n_examples]:
        print(f"\nSequence {seq_id}:")
        for pos, true, pred in errors[:3]:  # Show max 3 errors per sequence
        # Added padding handling after initial index errors
            context = X_test[seq_id][max(0,pos-2):min(len(X_test[seq_id]),pos+3)]
            print(f"Position {pos}: {X_test[seq_id][pos]}")
            print(f"True: {Config.ALL_TAGS[true]}, Pred: {Config.ALL_TAGS[pred]}")
            print("Context:", " ".join(context))

#  Results Discussion
def discuss_results(results):

    best_model = max(results, key=lambda x: x['f1'])
    worst_model = min(results, key=lambda x: x['f1'])

    print(f"\nBest Model: {best_model['model']} (F1: {best_model['f1']:.3f})")
    print(f"Worst Model: {worst_model['model']} (F1: {worst_model['f1']:.3f})")


def evaluate_model(name, y_true, y_pred):
    """Comprehensive model evaluation"""
    valid_pairs = [
        (true, pred) for true, pred in zip(y_true, y_pred)
        if len(true) == len(pred) and len(true) > 0
    ]

    if not valid_pairs:
        print(f"\n=== {name} Evaluation ===\nNo valid sequences")
        return {'model': name, 'f1': 0, 'precision': 0, 'recall': 0}

    y_true_valid, y_pred_valid = zip(*valid_pairs)
    y_true_decoded = [[Config.ALL_TAGS[t] for t in seq] for seq in y_true_valid]
    y_pred_decoded = [[Config.ALL_TAGS[p] for p in seq] for seq in y_pred_valid]

    print(f"\n=== {name} Evaluation ===")
    print(classification_report(y_true_decoded, y_pred_decoded))

    if Config.PLOT_RESULTS:
        try:
            plt.figure(figsize=(8,6))
            y_true_flat = list(itertools.chain(*y_true_decoded))
            y_pred_flat = list(itertools.chain(*y_pred_decoded))
            unique_tags = sorted(set(y_true_flat + y_pred_flat))
            cm = confusion_matrix(y_true_flat, y_pred_flat, labels=unique_tags)
            sns.heatmap(cm, annot=True, fmt='d', xticklabels=unique_tags,
                       yticklabels=unique_tags, cmap='Blues')
            plt.title(f"{name} Confusion Matrix")
            plt.xticks(rotation=45)
            plt.yticks(rotation=0)
            plt.show()
        except Exception as e:
            print(f"Confusion matrix error: {e}")

    return {
        'model': name,
        'f1': f1_score(y_true_decoded, y_pred_decoded, average='weighted'),
        'precision': precision_score(y_true_decoded, y_pred_decoded, average='weighted'),
        'recall': recall_score(y_true_decoded, y_pred_decoded, average='weighted')
    }
# Workflow
def main():
    """Complete end-to-end workflow"""
    # Load and analyze data
    X_train, pos_train, y_train, X_test, pos_test, y_test = load_data()
    dataset_stats = analyze_dataset(X_train, y_train, pos_train)

    # Train  with hyperparameter tuning and evaluate models
    results = []

    # HMM
    hmm_model = HMMModel(len(Config.TAG_MAP), Config.HMM_ITERATIONS)
    hmm_model.train(X_train, y_train)
    hmm_pred = hmm_model.predict(X_test)
    results.append(evaluate_model("HMM", y_test, hmm_pred))
    error_analysis(X_test, y_test, hmm_pred, "HMM")

    # CRF
    crf_model = CRFModel()
    crf_model.train(X_train, pos_train, y_train)
    crf_pred = crf_model.predict(X_test, pos_test)
    results.append(evaluate_model("CRF", y_test, crf_pred))
    error_analysis(X_test, y_test, crf_pred, "CRF")

    # SVM
    svm_model = SVMModel()
    svm_model.train(X_train, y_train)
    svm_pred = svm_model.predict(X_test)
    results.append(evaluate_model("SVM", y_test, svm_pred))
    error_analysis(X_test, y_test, svm_pred, "SVM")

    # Error analysis and comparison
    results_df = pd.DataFrame(results)
    print("\n Final Model Comparison ")
    print(results_df)

    if Config.PLOT_RESULTS:
        plt.figure(figsize=(8,4))
        results_df.set_index('model')[['f1', 'precision', 'recall']].plot(kind='bar')
        plt.title("Model Performance Comparison")
        plt.ylabel("Score")
        plt.ylim(0, 1)
        plt.tight_layout()
        plt.show()

    # Discussion
    discuss_results(results)

    # Save models
    joblib.dump({
        'hmm': hmm_model,
        'crf': crf_model,
        'svm': svm_model,
        'config': Config,
        'dataset_stats': dataset_stats
    }, 'token_classification_system.joblib')

if __name__ == "__main__":
    main()
```
